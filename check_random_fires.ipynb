```python
import os
import numpy as np
import matplotlib.pyplot as plt
import glob
import random
```

```python
#pathway to check
in_path = "/explore/nobackup/people/spotter5/cnn_mapping/nbac_training/l8_sent_collection2_proj_pos_subs_0_128"

#l8_sent_collection2_proj_pos_subs_0_128
#l8_sent_collection2_wgs_subs_0_128

#get list of all files ending in .npy

to_check = glob.glob(os.path.join(in_path, '*.npy'))

#random file
f = random.sample(to_check, 1)[0]

print(f)

#bad files which I need to check, 6658, 4758, 751, 10359, 6834, 4310, 1_3_8697

x = np.load(f)[:, :, 6]
y = np.load(f)[:, :, -1]


#setup plot, ax is plot 1, ax1 is plot 2 and cax is subplots
fig, (ax, ax1) = plt.subplots(ncols = 2, figsize = (15, 5),
                       gridspec_kw={"width_ratios":[1,1]})

#first plot
one = ax.imshow(x, cmap='RdBu_r')

#color bar for first plot
fig.colorbar(one, ax=ax, location='right', anchor=(0, 0.3), shrink=0.9)

#second plot
two = ax1.imshow(y)
fig.colorbar(two, ax=ax1, location='right', anchor=(0, 0.3), shrink=0.9)

plt.subplots_adjust(wspace=0.1, hspace=0, right=0.7)
```
